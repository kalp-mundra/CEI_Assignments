4.3 Eigenvalues & Eigenvectors

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[4, 1], [2, 3]], dtype=float)

# YOUR CODE HERE
eigenvalues, eigenvectors = np.linalg.eig(A)

print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors (columns):\n{eigenvectors}")

# Verify Av = λv for each eigenpair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    Av = A @ v
    lv = lam * v

    print(f"\nEigenpair {i+1}:")
    print("A @ v =", Av)
    print("λ * v =", lv)

    assert np.allclose(Av, lv), "Eigenvector equation failed!"

# Plot
fig, ax = plt.subplots(figsize=(6, 6))

for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v = eigenvectors[:, i]

    # Scale eigenvector by its eigenvalue
    scaled_v = lam * v

    ax.quiver(
        0, 0,
        scaled_v[0], scaled_v[1],
        angles='xy',
        scale_units='xy',
        scale=1,
        label=f'λ={lam:.2f}'
    )

ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Eigenvectors Scaled by Eigenvalues")
ax.axhline(0, color='k', lw=0.8)
ax.axvline(0, color='k', lw=0.8)
ax.grid(True)
ax.legend()

plt.tight_layout()
plt.show()

print("4.3 passed ✓")
```

### Geometric Interpretation of Eigenvalues and Eigenvectors

A matrix can be thought of as a transformation that changes vectors in space. When a matrix acts on a vector, it may rotate it, stretch it, compress it, or shear it.

An **eigenvector** is a special vector whose direction remains unchanged after the transformation. The vector may become longer, shorter, or reverse direction, but it still lies along the same line.

The corresponding **eigenvalue** tells us how much the eigenvector is scaled:

- If **λ > 1**, the vector is stretched (made longer).
- If **0 < λ < 1**, the vector is squished or compressed (made shorter).
- If **λ < 0**, the vector is flipped and scaled by |λ|.
- If **λ = 1**, the vector remains unchanged in length.

Mathematically:

A v = λ v

This means that applying the matrix A to the eigenvector v produces a new vector that is simply λ times the original vector.

Geometrically, eigenvectors represent the directions that are preserved by the transformation, while eigenvalues describe how much the transformation stretches or compresses vectors along those directions.

### SVD → PCA Connection

Singular Value Decomposition (SVD) decomposes a data matrix X into:

X = UΣVᵀ

where:

- **U** contains the left singular vectors. These describe how each data sample is represented in the transformed space.
- **S (Σ)** contains the singular values, which indicate the importance or strength of each component. Larger singular values correspond to directions that capture more variation in the data.
- **Vᵀ** contains the right singular vectors. These vectors represent the principal directions (axes) in feature space.

#### Relationship to PCA

PCA finds the directions along which the data varies the most. When SVD is applied to a centered data matrix:

X = UΣVᵀ

the rows of **Vᵀ** (or equivalently, the columns of **V**) are the **principal components**.

- The **principal component directions** are stored in **V**.
- The **amount of variance explained** by each component is related to the squared singular values (S²).
- Components with larger singular values capture more information about the data.

#### Which Matrix Contains the Directions of Maximum Variance?

The matrix **V** (or equivalently the rows of **Vᵀ**) contains the directions of maximum variance. These are the principal component axes found by PCA.

In summary:

- **U** → representation of samples in the new coordinate system.
- **S** → importance (strength/variance) of each component.
- **Vᵀ** → principal component directions.
- **V** contains the actual directions of maximum variance used in PCA.

5.3 Error Metrics

```python
import numpy as np

y_true = np.array([3.0, 5.0, 2.5, 7.0, 4.5, 6.0, 1.5, 8.0])
y_pred = np.array([2.8, 5.2, 2.1, 7.5, 4.0, 6.3, 2.0, 7.8])

n, p = len(y_true), 2

# YOUR CODE HERE

# MAE
mae = np.mean(np.abs(y_true - y_pred))

# MSE
mse = np.mean((y_true - y_pred) ** 2)

# RMSE
rmse = np.sqrt(mse)

# R² Score
ss_res = np.sum((y_true - y_pred) ** 2)
ss_tot = np.sum((y_true - np.mean(y_true)) ** 2)

r2 = 1 - (ss_res / ss_tot)

# Adjusted R²
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print(f"MAE:        {mae:.4f}")
print(f"MSE:        {mse:.4f}")
print(f"RMSE:       {rmse:.4f}")
print(f"R²:         {r2:.4f}")
print(f"Adj. R²:    {adj_r2:.4f}")
```

5.4 Distribution Testing & Stationarity

```python
from scipy import stats
from statsmodels.tsa.stattools import adfuller
import numpy as np

np.random.seed(0)

s1 = np.random.normal(0, 1, 200)
s2 = np.random.exponential(1, 200)

# KS test — compare each against N(0,1)
ks_s1 = stats.kstest(s1, 'norm')
ks_s2 = stats.kstest(s2, 'norm')

print(f"KS test s1 (normal):      stat={ks_s1.statistic:.4f}, p={ks_s1.pvalue:.4f}")
print(f"KS test s2 (exponential): stat={ks_s2.statistic:.4f}, p={ks_s2.pvalue:.4f}")

# ADF test on non-stationary series
t = np.arange(200)
ts = 0.05 * t + np.random.normal(0, 1, 200)

adf_result = adfuller(ts)

print(f"\nADF statistic: {adf_result[0]:.4f}")
print(f"ADF p-value:   {adf_result[1]:.4f}")
print(f"Stationary?    {adf_result[1] < 0.05}")

# Difference once and re-run
ts_diff = np.diff(ts)

adf_diff = adfuller(ts_diff)

print(
    f"\nAfter differencing — p-value: {adf_diff[1]:.4f}, "
    f"Stationary? {adf_diff[1] < 0.05}"
)
```

5.5 Model Monitoring Concepts

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_psi(expected, actual, bins=10):
    """
    PSI = sum((actual_pct - expected_pct) * ln(actual_pct / expected_pct))
    Add epsilon=1e-10 to avoid log(0).
    """
    epsilon = 1e-10

    # Create bins based on expected distribution
    breakpoints = np.linspace(0, 100, bins + 1)
    bin_edges = np.percentile(expected, breakpoints)

    expected_counts, _ = np.histogram(expected, bins=bin_edges)
    actual_counts, _ = np.histogram(actual, bins=bin_edges)

    expected_pct = expected_counts / len(expected)
    actual_pct = actual_counts / len(actual)

    psi = np.sum(
        (actual_pct - expected_pct) *
        np.log((actual_pct + epsilon) / (expected_pct + epsilon))
    )

    return psi

np.random.seed(1)

train_dist = np.random.normal(50, 10, 1000)    # training distribution
drift_dist = np.random.normal(65, 12, 1000)    # shifted production distribution

psi_value = compute_psi(train_dist, drift_dist)

print(f"PSI: {psi_value:.4f}")
print(
    f"Shift severity: "
    f"{'Major' if psi_value > 0.2 else 'Minor' if psi_value > 0.1 else 'Stable'}"
)

# Plot both distributions
fig, ax = plt.subplots()

# YOUR CODE HERE
ax.hist(train_dist, bins=30, alpha=0.6, density=True, label='Training Distribution')
ax.hist(drift_dist, bins=30, alpha=0.6, density=True, label='Production Distribution')

ax.legend()
ax.set_title('Distribution Shift Visualization')

plt.tight_layout()
plt.show()
```

### Concept Drift vs Covariate Drift & PSI Thresholds (5.5)

**Concept Drift:**  
The relationship between input features and the target variable changes over time. In other words, the patterns the model learned are no longer valid.

**Covariate Drift:**  
The distribution of input features changes over time, even if the relationship between features and the target remains the same.

**PSI < 0.1:**  
No significant shift. The feature distribution is considered stable.

**PSI 0.1–0.2:**  
Moderate or minor shift. Monitoring is recommended.

**PSI > 0.2:**  
Major shift. The feature distribution has changed substantially and may impact model performance.

**Retraining Trigger Example:**  
Retrain the model when a key feature's PSI exceeds 0.2 for multiple monitoring periods or when model performance metrics (e.g., accuracy, RMSE, AUC) degrade beyond an acceptable threshold.

6.3 Bayes' Theorem

```python
p_spam            = 0.30
p_free_given_spam = 0.80
p_free_given_ham  = 0.05
p_ham             = 1 - p_spam

# P(free) — law of total probability
p_free = (
    p_free_given_spam * p_spam +
    p_free_given_ham * p_ham
)

# Bayes
p_spam_given_free = (
    p_free_given_spam * p_spam
) / p_free

print(f"P('free'): {p_free:.4f}")
print(f"P(Spam | 'free'): {p_spam_given_free:.4f}")

def naive_bayes_predict(prior_spam, p_word_given_spam, p_word_given_ham):
    p_ham = 1 - prior_spam

    evidence = (
        p_word_given_spam * prior_spam +
        p_word_given_ham * p_ham
    )

    posterior = (
        p_word_given_spam * prior_spam
    ) / evidence

    return posterior

pred = naive_bayes_predict(0.30, 0.80, 0.05)

assert abs(pred - p_spam_given_free) < 1e-9

print(f"naive_bayes_predict: {pred:.4f}")
print("6.3 passed ✓")
```

## Bayes Term Mapping (6.3)

| Term | Formula | In this example |
|--------|---------|----------------|
| Prior | P(Spam) | Probability that an email is spam before seeing any words = 0.30 |
| Likelihood | P(word \| Spam) | Probability of seeing the word "free" given that the email is spam = 0.80 |
| Evidence | P(word) | Overall probability of seeing the word "free" in any email = 0.275 |
| Posterior | P(Spam \| word) | Probability that an email is spam after observing the word "free" = 0.8727 |

### Interpretation

- **Prior:** Our initial belief about an email being spam.
- **Likelihood:** How likely the word "free" appears in spam emails.
- **Evidence:** How common the word "free" is across all emails.
- **Posterior:** Our updated belief that the email is spam after seeing the word "free".

6.4 Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(7)

population = np.random.exponential(scale=1.0, size=100_000)

n_samples = 5000
sample_size = 30

# YOUR CODE HERE — array of 5000 sample means
sample_means = np.array([
    np.random.choice(population, size=sample_size, replace=True).mean()
    for _ in range(n_samples)
])

pop_mean = population.mean()
pop_std = population.std()

# YOUR CODE HERE
clt_std = pop_std / np.sqrt(sample_size)

print(f"Population mean: {pop_mean:.4f}, Population std: {pop_std:.4f}")
print(f"Sample means mean: {sample_means.mean():.4f}, std: {sample_means.std():.4f}")
print(f"CLT predicted std: {clt_std:.4f}")

# Plot
fig, ax = plt.subplots()

# YOUR CODE HERE
ax.hist(
    sample_means,
    bins=40,
    density=True,
    alpha=0.7,
    label='Sample Means'
)

x = np.linspace(
    sample_means.min(),
    sample_means.max(),
    300
)

ax.plot(
    x,
    stats.norm.pdf(x, loc=pop_mean, scale=clt_std),
    linewidth=2,
    label='CLT Normal Approximation'
)

ax.legend()
ax.set_title('CLT: Distribution of Sample Means')

plt.tight_layout()
plt.show()

# KS test
ks_result = stats.kstest(
    sample_means,
    'norm',
    args=(pop_mean, clt_std)
)

print(
    f"KS test p-value: {ks_result.pvalue:.4f} "
    f"→ Approximately normal? {ks_result.pvalue > 0.05}"
)
```

## CLT Reflection (6.4)

The Central Limit Theorem (CLT) shows that the distribution of sample means becomes approximately normal as the sample size increases, even when the original population is not normally distributed. In this example, the population is exponential (highly skewed), yet the sample means form an approximately normal distribution.

This result is important because many statistical and machine learning methods rely on normality assumptions. The CLT justifies the use of classical statistical procedures such as confidence intervals, hypothesis tests, t-tests, and linear regression inference when the sample size is sufficiently large.
